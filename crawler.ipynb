### Read data

```python
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
```

```python
#read csv rows
import csv
with open('Greek_Parliament_Proceedings_1989_2020.csv', encoding="utf8") as file:
    reader = csv.reader(file, delimiter=',', quotechar='"',)
    rows = []
    counter = 0
    for row in reader:
        rows.append(row)
        counter+=1
        #print(counter)
        if(counter==33081):
            break
        
labels = rows[0:1]     
rows = rows[1:]
print(len(labels[0]))
print((labels[0]))
```

```python
#read stopwords
with open('stopwords.txt', encoding='utf-8') as file:
    lines = [line.rstrip() for line in file]
print(lines)
```

```python
#remoce punctuation
import string

punctuation = string.punctuation.replace(':','')
for row in rows:
    row[10] = row[10].translate(str.maketrans('','',punctuation))
```

```python
#create index 
index = {}
for i, row in enumerate(rows):
    words = [w for w in word_tokenize(row[10]) if w not in lines]
    unique_words = set(words)
    for word in unique_words:
        if word in index.keys():
            index[word][1].append([i, words.count(word)])
            index[word][0] = index[word][0] + 1
        else:
            index.update({word:[1,[[i,words.count(word)]]]})
```
